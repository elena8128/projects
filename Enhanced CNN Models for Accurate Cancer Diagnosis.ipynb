# Task and Data Summary

This project is centered around a Kaggle competition that presents a challenging task: to develop a convolutional neural network (CNN) capable of identifying metastatic cancer within histopathological images of lymph nodes. The goal is to build a model that can effectively analyze small image patches, discerning whether they contain cancerous cells or not. This task is particularly important in the field of medical diagnostics, where accurate and timely detection of cancer can significantly impact patient outcomes.

### Dataset Composition

The dataset used in this project is divided into two main components, each crucial for the development and testing of the CNN model. The first component is a large collection of histopathological images, all of which are derived from lymph node tissue samples. Each image in this dataset is saved in a `.tif` format, with dimensions of 96x96 pixels and three color channels (RGB). The dataset is structured into two subsets: a training set and a test set. The training set is used to develop and refine the CNN model, while the test set is reserved for evaluating the model's performance after training.

These images are not merely random samples; they have been carefully curated and labeled to facilitate the task of cancer detection. The images in the training set come with associated labels that indicate whether or not there is metastatic cancer present in the central 32x32 pixel region of the image. This labeling is binary: a label of `1` signifies the presence of cancerous tissue, while a label of `0` indicates the absence of such tissue. The labels are essential for training the CNN, as they provide the ground truth needed for supervised learning. The training set contains a total of 220,025 images, which is a substantial amount of data, making it possible to train a robust model. ### Research Focus and Approach

The primary focus of this research is to leverage the power of CNNs to detect cancerous tissues in histopathological images. CNNs are a class of deep learning models that have proven to be highly effective in image classification tasks, particularly in the medical field. The use of CNNs in this project is motivated by their ability to automatically learn hierarchical features from raw image data, which can then be used to distinguish between cancerous and non-cancerous tissues. This is particularly important given the subtle differences that often exist between healthy and diseased tissues, differences that may not be easily discernible by the human eye.

To develop the CNN model, a series of steps were undertaken, starting with the preprocessing of the image data. Preprocessing is a critical step in any image classification task, as it ensures that the data fed into the model is of high quality and suitable for learning. In this project, preprocessing involved resizing the images to ensure uniformity in dimensions, normalizing the pixel values to a range suitable for the activation functions used in the CNN, and augmenting the data to increase the diversity of the training set. Data augmentation is particularly important in medical image analysis, as it helps to simulate variations that might occur in real-world scenarios, such as differences in lighting, orientation, and scale.

Following preprocessing, the next step was to design the architecture of the CNN model. The architecture of a CNN is composed of several layers, each of which plays a specific role in the feature extraction and classification process. The initial layers of the CNN are typically convolutional layers, which apply filters to the input images to detect various features such as edges, textures, and shapes. These features are then passed through pooling layers, which reduce the spatial dimensions of the feature maps, making the model more computationally efficient and reducing the risk of overfitting.

In this project, the CNN model was designed with multiple convolutional layers, each followed by a pooling layer. The number of filters in the convolutional layers was gradually increased to allow the model to capture more complex features as the data progresses through the network. Batch normalization was also applied after each convolutional layer to stabilize the learning process and improve convergence. Additionally, dropout layers were included to prevent overfitting by randomly deactivating a certain percentage of neurons during training, thus encouraging the model to learn more robust features.

The final layers of the CNN are fully connected (dense) layers, which are responsible for making the final classification based on the features extracted by the earlier layers. In this project, the output layer of the CNN used a sigmoid activation function, which is appropriate for binary classification tasks. The model was trained using a binary cross-entropy loss function, which is commonly used for binary classification problems, and the Adam optimizer, which is known for its efficiency in training deep learning models.

# Neural Network Design

Initially, the 'train' dataset will be partitioned into two subsets: a training set and a validation set, with an 80:20 ratio. The model will be trained using the training subset and evaluated on the validation subset to assess its performance.

```python
train, val = train_test_split(df, test_size=0.2, stratify=df['label'], random_state=42)
```

For our convolutional neural network we will be using a sequential model with the simple structure of convolution-convolution-MaxPool.  We will create a basic, barebones model with default hyperparameters as well as only one hidden layer.  This is to test how the model to interact with images.

```python
# Define model parameters
filter_1 = 32
filter_size = (3, 3)
img_width = img.width  # Width of the input image (96 pixels)
img_height = img.height  # Height of the input image (96 pixels)
img_depth = 3  # Number of color channels (RGB)
pool_size = (2, 2)

# Initialize the sequential model
model = Sequential()

# Add the first convolutional layer with ReLU activation
model.add(Conv2D(filter_1, filter_size, activation='relu', input_shape=(img_width, img_height, img_depth)))

# Add a second convolutional layer with ReLU activation
model.add(Conv2D(filter_1, filter_size, activation='relu'))

# Add a max-pooling layer to reduce spatial dimensions
model.add(MaxPooling2D(pool_size))

# Flatten the output from the convolutional layers
model.add(Flatten())

# Add a dense layer with ReLU activation
model.add(Dense(filter_1 * 2, activation='relu'))

# Add the output layer with a sigmoid activation function for binary classification
model.add(Dense(1, activation='sigmoid'))

# Compile the model with the Adam optimizer and binary cross-entropy loss function
model.compile(Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])
```

As I struggled with figuring out how to input images into my Sequential CNN model, I resorted to referencing the code in the notebook <a href= "https://www.kaggle.com/code/fadhli/starter-code-keras-resnet50-0-9275-lb/notebook">Starter Code Keras</a> to view how the ImageDataGenerator function was utilized.

```python
# Data augmentation and normalization for the training set
train_datagen = ImageDataGenerator(
    preprocessing_function=lambda x: (x - x.mean()) / x.std() if x.std() > 0 else x,
    horizontal_flip=True,
    vertical_flip=True
)

# Normalization for the validation set (no augmentation)
test_datagen = ImageDataGenerator(
    preprocessing_function=lambda x: (x - x.mean()) / x.std() if x.std() > 0 else x
)

# Generator for the training data with augmentation and shuffling
train_generator = train_datagen.flow_from_dataframe(
    dataframe=train,
    directory=path + 'train/',
    x_col='file',
    y_col='label',
    has_ext=False,
    batch_size=32,
    seed=2018,
    shuffle=True,
    class_mode='binary',
    target_size=(96, 96),
    validate_filenames=False
)

# Generator for the validation data without augmentation, no shuffling
valid_generator = test_datagen.flow_from_dataframe(
    dataframe=val,
    directory=path + 'train/',
    x_col='file',
    y_col='label',
    has_ext=False,
    batch_size=32,
    seed=2018,
    shuffle=False,
    class_mode='binary',
    target_size=(96, 96),
    validate_filenames=False
)

# Calculate steps per epoch for training and validation
STEP_SIZE_TRAIN = train_generator.n // train_generator.batch_size
STEP_SIZE_VALID = valid_generator.n // valid_generator.batch_size

# Early stopping to prevent overfitting, with patience of 2 epochs
earlystopper = EarlyStopping(monitor='val_loss', patience=2, verbose=1, restore_best_weights=True)

# Reduce learning rate when a metric has stopped improving, with patience of 1 epoch
reducel = ReduceLROnPlateau(monitor='val_loss', patience=1, verbose=1, factor=0.1)
```

```python
history = model.fit(train_generator, steps_per_epoch=STEP_SIZE_TRAIN,
                    validation_data=valid_generator,
                    validation_steps=STEP_SIZE_VALID,
                    epochs=5,
                   callbacks=[reducel, earlystopper])
```

```python
plt.plot(history.history['val_accuracy'], label = 'validation accuracy')
plt.plot(history.history['accuracy'], label = 'training accuracy')
plt.legend()
plt.show()
```

Our accuracy plot reveals that the base model can achieve a solid training accuracy of over 0.84. However, the training accuracy is noticeably higher than the validation accuracy, suggesting a degree of overfitting. Since this is just our initial model, we can begin to refine hyperparameters and modify the architecture to address this.

## 1. Implementing Dropout Layers

The first modification involves adding dropout layers following the hidden layers and before the output layer. This approach is intended to enhance regularization and mitigate overfitting. While the overall performance metrics of the model remain largely consistent, the gap between training and validation accuracy has notably diminished, indicating improved generalization to new data.

```python
# Define model parameters
filter_1 = 32
filter_size = (3, 3)
img_width = img.width  # Width of the input image (96 pixels)
img_height = img.height  # Height of the input image (96 pixels)
img_depth = 3  # Number of color channels (RGB)
dropout_conv = 0.3  # Dropout rate for convolutional layers
dropout_dense = 0.5  # Dropout rate for dense layers
pool_size = (2, 2)  # Pooling size for MaxPooling layers

# Initialize the sequential model
model_1 = Sequential()

# Add the first convolutional layer with ReLU activation
model_1.add(Conv2D(filter_1, filter_size, activation='relu', input_shape=(img_width, img_height, img_depth)))

# Add a second convolutional layer with ReLU activation
model_1.add(Conv2D(filter_1, filter_size, activation='relu'))

# Add a max-pooling layer to down-sample the input
model_1.add(MaxPooling2D(pool_size))

# Apply dropout to reduce overfitting
model_1.add(Dropout(dropout_conv))

# Flatten the output from the convolutional layers for the dense layer
model_1.add(Flatten())

# Add a dense layer with ReLU activation
model_1.add(Dense(filter_1 * 2, activation='relu'))

# Apply dropout before the output layer
model_1.add(Dropout(dropout_dense))

# Add the output layer with a sigmoid activation function for binary classification
model_1.add(Dense(1, activation='sigmoid'))

# Compile the model using the Adam optimizer and binary cross-entropy loss function
model_1.compile(Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history_1 = model_1.fit(train_generator, steps_per_epoch=STEP_SIZE_TRAIN,
                    validation_data=valid_generator,
                    validation_steps=STEP_SIZE_VALID,
                    epochs=5,
                   callbacks=[reducel, earlystopper])
```

```python
plt.plot(history_1.history['val_accuracy'], label = 'validation accuracy')
plt.plot(history_1.history['accuracy'], label = 'training accuracy')
plt.legend()
plt.show()
```

## 2. Incorporating Additional Hidden Layers

The base model is limited to just one hidden layer, which restricts its capacity to learn complex patterns from the data. To enhance the model's learning capability, we introduced two additional hidden layers. Each new layer was designed with a filter size that is double that of the previous layer, allowing the model to capture more intricate features.

The results showed that, although the computation time increased, this adjustment led to improvements in both training and validation accuracy, indicating a better overall performance and enhanced ability to generalize from the training data.

```python
filter_1 = 32
filter_2 = 64
filter_3 = 128
filter_size = ( 3, 3)
img_width = img.width # 96
img_height = img.height # 96
img_depth = 3 # RGB
dropout_conv = 0.3
dropout_dense = 0.5
pool_size = ( 2, 2)

model_2 = Sequential()

model_2.add(Conv2D(filter_1, filter_size, activation = 'relu', input_shape = (img_width, img_height, img_depth)))
model_2.add(Conv2D(filter_1, filter_size, activation = 'relu'))
model_2.add(MaxPooling2D(pool_size))
model_2.add(Dropout(dropout_conv))

model_2.add(Conv2D(filter_2, filter_size, activation = 'relu'))
model_2.add(Conv2D(filter_2, filter_size, activation = 'relu'))
model_2.add(MaxPooling2D(pool_size))
model_2.add(Dropout(dropout_conv))

model_2.add(Conv2D(filter_3, filter_size, activation = 'relu'))
model_2.add(Conv2D(filter_3, filter_size, activation = 'relu'))
model_2.add(MaxPooling2D(pool_size))
model_2.add(Dropout(dropout_conv))

model_2.add(Flatten())
model_2.add(Dense(256, activation = 'relu'))
model_2.add(Dropout(dropout_dense))
model_2.add(Dense(1, activation = 'sigmoid'))

model_2.compile(Adam(learning_rate = 0.001), loss='binary_crossentropy', metrics=['accuracy'])
```

```python
filter_1 = 32
filter_2 = 64
filter_3 = 128
filter_size = ( 3, 3)
img_width = img.width # 96
img_height = img.height # 96
img_depth = 3 # RGB
dropout_conv = 0.3
dropout_dense = 0.5
pool_size = ( 2, 2)

model_5 = Sequential()

model_5.add(Conv2D(filter_1, filter_size, activation = 'relu', input_shape = (img_width, img_height, img_depth)))
model_5.add(Conv2D(filter_1, filter_size, activation = 'relu'))
model_5.add(MaxPooling2D(pool_size))
model_5.add(Dropout(dropout_conv))

model_5.add(Conv2D(filter_2, filter_size, activation = 'relu'))
model_5.add(Conv2D(filter_2, filter_size, activation = 'relu'))
model_5.add(MaxPooling2D(pool_size))
model_5.add(Dropout(dropout_conv))

model_5.add(Conv2D(filter_3, filter_size, activation = 'relu'))
model_5.add(Conv2D(filter_3, filter_size, activation = 'relu'))
model_5.add(MaxPooling2D(pool_size))
model_5.add(Dropout(dropout_conv))

model_5.add(Flatten())
model_5.add(Dense(256, activation = 'relu'))
model_5.add(Dropout(dropout_dense))
model_5.add(Dense(1, activation = 'sigmoid'))

model_5.compile(Adam(learning_rate = 0.0001), loss='binary_crossentropy', metrics=['accuracy'])
```

<h3>6. Introducing batch normalization</h3><a name="df"></a>

Similar to dropout layers, batch normalization is another regularization method we can employ.  From the results, both the training accuracy and the validation accuracy have increased slightly over the previous methods.  Thus this is a change we can keep.

```python
filter_1 = 32
filter_2 = 64
filter_3 = 128
filter_size = ( 3, 3)
img_width = img.width # 96
img_height = img.height # 96
img_depth = 3 # RGB
dropout_conv = 0.3
dropout_dense = 0.5
pool_size = ( 2, 2)

model_6 = Sequential()

model_6.add(Conv2D(filter_1, filter_size, activation = 'relu', input_shape = (img_width, img_height, img_depth)))
model_6.add(BatchNormalization())
model_6.add(Conv2D(filter_1, filter_size, activation = 'relu'))
model_6.add(BatchNormalization())
model_6.add(MaxPooling2D(pool_size))
model_6.add(Dropout(dropout_conv))

model_6.add(Conv2D(filter_2, filter_size, activation = 'relu'))
model_6.add(BatchNormalization())
model_6.add(Conv2D(filter_2, filter_size, activation = 'relu'))
model_6.add(BatchNormalization())
model_6.add(MaxPooling2D(pool_size))
model_6.add(Dropout(dropout_conv))

model_6.add(Conv2D(filter_3, filter_size, activation = 'relu'))
model_6.add(BatchNormalization())
model_6.add(Conv2D(filter_3, filter_size, activation = 'relu'))
model_6.add(BatchNormalization())
model_6.add(MaxPooling2D(pool_size))
model_6.add(Dropout(dropout_conv))

model_6.add(Flatten())
model_6.add(Dense(256, activation = 'relu'))
model_6.add(BatchNormalization())
model_6.add(Dropout(dropout_dense))
model_6.add(Dense(1, activation = 'sigmoid'))

model_6.compile(Adam(learning_rate = 0.001), loss='binary_crossentropy', metrics=['accuracy'])
```

<h3>7. Changing the output activation function</h3><a name="dg"></a>

For our final hyperparameter tuning, we will try changing the activation function of our output layer from a sigmoid activation to a softmax activation.  However as seen from our results, a model with a softmax activation function proved to be the worst out of all our current models.  This is most likely due to sigmoid being preferable for binary outputs while softmax is more geared for numerous categories.

```python
filter_1 = 32
filter_2 = 64
filter_3 = 128
filter_size = ( 3, 3)
img_width = img.width # 96
img_height = img.height # 96
img_depth = 3 # RGB
dropout_conv = 0.3
dropout_dense = 0.5
pool_size = ( 2, 2)

model_7 = Sequential()

model_7.add(Conv2D(filter_1, filter_size, activation = 'relu', input_shape = (img_width, img_height, img_depth)))
model_7.add(BatchNormalization())
model_7.add(Conv2D(filter_1, filter_size, activation = 'relu'))
model_7.add(BatchNormalization())
model_7.add(MaxPooling2D(pool_size))
model_7.add(Dropout(dropout_conv))

model_7.add(Conv2D(filter_2, filter_size, activation = 'relu'))
model_7.add(BatchNormalization())
model_7.add(Conv2D(filter_2, filter_size, activation = 'relu'))
model_7.add(BatchNormalization())
model_7.add(MaxPooling2D(pool_size))
model_7.add(Dropout(dropout_conv))

model_7.add(Conv2D(filter_3, filter_size, activation = 'relu'))
model_7.add(BatchNormalization())
model_7.add(Conv2D(filter_3, filter_size, activation = 'relu'))
model_7.add(BatchNormalization())
model_7.add(MaxPooling2D(pool_size))
model_7.add(Dropout(dropout_conv))

model_7.add(Flatten())
model_7.add(Dense(256, activation = 'relu'))
model_7.add(BatchNormalization())
model_7.add(Dropout(dropout_dense))
model_7.add(Dense(1, activation = 'softmax'))

model_7.compile(Adam(learning_rate = 0.001), loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history_7 = model_7.fit(train_generator, steps_per_epoch=STEP_SIZE_TRAIN,
                    validation_data=valid_generator,
                    validation_steps=STEP_SIZE_VALID,
                    epochs=5,
                   callbacks=[reducel, earlystopper])
```
